```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
spam_data=pd.read_csv('/content/mail_data.csv')
```

```python
spam_data.head()
```

```python
spam_data.shape
```

```python
spam_data.describe()
```

```python
spam_data['Category'].value_counts()
```

```python
spam_data.isnull().sum()
```

```python
# Label Encoding
spam_data.loc[spam_data['Category']=='spam','Category',]=0
spam_data.loc[spam_data['Category']=='ham',['Category'],]=1
```

```python
spam_data.head()
```

```python
# separating the data as text and categotry
x=spam_data['Message']
y=spam_data['Category']
```

```python
x,y
```

```python
# Splitting the data into train and test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
# Feature Extraction
# changing the text to small letter
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_feature=feature_extraction.transform(x_test)

# Convert y_train and y_test as values
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
# Training the model
model = LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
# Evaluating the model
# prediction on training data
prediction_training=model.predict(x_train_features)
acc_training=accuracy_score(y_train,prediction_training)
print('Accuracy=',acc_training)
```

```python
# Evaluating the model
# prediction on test data
prediction_test=model.predict(x_test_feature)
acc_test=accuracy_score(y_test,prediction_test)
print('Accuracy=',acc_test)
```

```python
# Making a predictive model
input=["I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k? I've cried enough today."]
# Convert text to Feature vectors
input_data_feature=feature_extraction.transform(input)
# making prediction

prediction=model.predict(input_data_feature)
print(prediction)
```
